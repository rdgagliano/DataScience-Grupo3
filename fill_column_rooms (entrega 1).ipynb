### Método que a partir de un text busca palabras claves para contar los ambientes

##### Cuanta los dormitorios


```python
def getDormitoryCount(text):
    count = 0;    
    result = '';
    
    m = regex.search('\d dorm', text.lower())
    
    if(m is not None):
        result = m.group(0)
    
    if(result == ''):                        
        m = regex.search('un dorm', text.lower())    
        if(m is not None):
            return 1
        
        m = regex.search('dos dorm', text.lower())    
        if(m is not None):
            return 2
        
        m = regex.search('tres dorm', text.lower())    
        if(m is not None):
            return 3
        
        m = regex.search('cuatro dorm', text.lower())    
        if(m is not None):
            return 4
        
        m = regex.search('cinco dorm', text.lower())    
        if(m is not None):
            return 5                
        
        m = regex.search('dorm', text.lower())    
        if(m is not None):
            return 1
    else:
        m = regex.search('\d', result.lower())
        
        if(m is not None):
            result = m.group(0)
        
        if(result != ''):
            count = int(result)
    
    return count;            
```

##### Cuenta living y comedor

```python
def getLivingCount(text):
    count = 0;    
        
    m = regex.search('liv\w com', text.lower())
        
    if(m is not None):
        return  1;
    
    m = regex.search('liv\w coc\w com', text.lower())
    
    if(m is not None):
        return  1;
    
    if(count == 0):
        m = regex.search('living', text.lower())
        
        if(m is not None):
            count += 1;
        
        m = regex.search('comedor', text.lower())
        
        if(m is not None):
            count += 1;
            
    return count;            
```

##### Cuenta cocina

```python
def getKitchenCount(text):
    count = 0;    
        
    m = regex.search('cocina separada', text.lower())
    
    if(m is not None):
        count = 1;
        
    return count;            
```

### Metodo que a partir de un texto retorna la cantidad de ambientes

```python
def getRoomsNumberFromText(text):
    roomsCount = 0;
    result = '';
    
    #para cuando el texto tiene números, ej: '2 ambientes'
    m = regex.search('\d', text)
    
    if(m is not None):
        result = m.group(0)
    
    if(result == ''):
        #no funcionó el parseo por números, se busca por palabras: monoambiente, dos ambientes, tres ambientes
        if('tienda' in text.lower()): roomsCount = 1
        if('local' in text.lower()): roomsCount = 1
        if('oficina' in text.lower()): roomsCount = 1
        if('monoambiente' in text.lower()): roomsCount = 1
        if('mono ambiente' in text.lower()): roomsCount = 1
        if('ambiente unico' in text.lower()): roomsCount = 1
        if('casa' in text.lower()): roomsCount = 4
        if('un amb' in text.lower()): roomsCount = 1
        if('uno' in text.lower()): roomsCount = 1
        if('dos amb' in text.lower()): roomsCount = 2
        if('tres amb' in text.lower()): roomsCount = 3
        if('cuatro amb' in text.lower()): roomsCount = 4
        if('cinco amb' in text.lower()): roomsCount = 5
        if('seis amb' in text.lower()): roomsCount = 6
        if('siete amb' in text.lower()): roomsCount = 7
        if('ocho amb' in text.lower()): roomsCount = 8
        if('nueve amb' in text.lower()): roomsCount = 9
        if('diez amb' in text.lower()): roomsCount = 10
    else:
        roomsCount = int(result)
    
    return roomsCount
    
```

### Método que a partir de un dato Description del dataset, retorna la cantidad de ambientes

#### Este método llama getRoomsNumberFromText() para obtener el número de ambientes

```python
def getRoomsFromDescription(description):
    
    #array con expresiones regulares de búsqueda del dato 'cantidad de ambientes'
    #se van a buscar estos string por separado en el parámetro description
    roomsNumberRegexs = ['\d amb', 'monoambiente', 'mono ambiente', 'un amb', 'dos amb', 'tres amb', 'cuatro amb', 'cinco amb', 'seis amb', 'siete amb', 'ocho amb', 'local', 'tienda', 'oficina', 'casa']
    
    roomsNumber = ''
    
    #primer búsqueda, que se nombre en la description la cantidad de ambientes
    #por cada regex en roomsNumberRegexs
    for expr in roomsNumberRegexs:
        #busca si la expresion regular está en description
        m = regex.search(expr, description.lower())
        
        if(m is not None):
            roomsNumber = m.group(0)
            break #como ya se encontró, no se sigue buscando, se sale del ciclo
                        
    if(roomsNumber != ''):        
        #segunda búsqueda, si se encontró el texto con los ambientes, se obtiene el número de ambientes
        return getRoomsNumberFromText(roomsNumber)
    else:
        return getRoomsCountByKeyWords(description)
```

### Método para actualizar la cantidad de ambientes, según la description

#### Este método llama a getRoomsFromDescription(), para obtener la cantidad de rooms desde la description y con eso actualizar el dato Rooms

```python
def updateRoomsQuantityFromDescription(dataSetRow):
    if(math.isnan(dataSetRow.rooms)):
        try:
            if(dataSetRow.description is not None):

                roomsCount = 0

                roomsCount = getRoomsFromDescription(dataSetRow.description)

                if(roomsCount is not None):
                    if(roomsCount > 0):
                        dataSetRow.rooms = roomsCount

            return dataSetRow
        except:
            print('No se pudo obtener el room para index: ' +  str(dataSetRow[0]))
    else:
        #el campo rooms ya tiene un valor, no hacer nada
        return dataSetRow
```

### Método para actualizar la cantidad de ambientes, según el title

```python
def updateRoomsQuantityFromTitle(dataSetRow):
    if(math.isnan(dataSetRow.rooms)):
        try:
            if(dataSetRow.title is not None):

                roomsCount = 0

                roomsCount = getRoomsFromDescription(dataSetRow.title)

                if(roomsCount is not None):
                    if(roomsCount > 0):
                        dataSetRow.rooms = roomsCount

            return dataSetRow
        except:
            print('No se pudo obtener el room para index: ' +  str(dataSetRow[0]))
    else:
        #el campo room ya tiene un valor, no hacer nada
        return dataSetRow
```

## Metodo para actualizar los datos a partir de la description y del title

```python
def updateRoomsFromRowData(dataSetRow):
    updatedDataSetRow = updateRoomsQuantityFromDescription(dataSetRow)
    
    if(updatedDataSetRow is not None):
        if(math.isnan(updatedDataSetRow.rooms)):
            updatedDataSetRow = updateRoomsQuantityFromTitle(dataSetRow)
    
    return updatedDataSetRow
```
