```python
import pandas as pd
import numpy as np
from IPython.display import display
data = pd.read_csv('properatti.csv', dtype={'all': str})
```

#### Armo un subset sólo con las columnas 'place_name' y 'lat'

```python
DataLat=data[['place_name', 'lat']]
DataLat
```

#### Armo un subset con las mismas dos columnas, pero me quedo solo con los valores de 'lat' completos

```python
DataLatNotNan=DataLat.dropna()
DataLatNotNan
```

#### Saco la media de las 'lat' por grupo de valores de la columna 'place_name'
#### Por ejemplo toma toda la localidad Abasto con lat NotNaN y promedia su valor de Latitud

```python
MediaLatbyLoc=DataLatNotNan.groupby(['place_name']).mean()
MediaLatbyLoc
```

#### Armo un subset sólo con las columnas 'place_name' y 'lon'

```python
DataLon=data[['place_name', 'lon']]
DataLon
```

#### Armo un subset con las mismas dos columnas, pero me quedo solo con los valores de 'lon' completos

```python
DataLonNotNan=DataLon.dropna()
DataLonNotNan
```

#### Saco la media de las 'lon' por grupo de valores de la columna 'place_name'
#### Por ejemplo toma toda la localidad Abasto con lon NotNaN y promedia su valor de Longitud

```python
MediaLonbyLoc=DataLonNotNan.groupby(['place_name']).mean()
MediaLonbyLoc
```

#### Armo un dataframe nuevo con método merge para unir las columnas 'lat' y 'lon'

```python
dflatlon=pd.merge(MediaLatbyLoc, MediaLonbyLoc, left_on='place_name', right_on='place_name')
dflatlon
```

#### Reindexamos para place_name vuelva a ser una variable

```python
dflatlon.reset_index(inplace=True)
dflatlon
```

```python
#### Uno el dataframe con las 'lat' y las 'lon' que promedié junto con el data original
```

```python
NuevoDF=pd.merge(dflatlon,data, left_on='place_name', right_on='place_name')
NuevoDF
```
