### Al parecer la primer columna corresponde a una campo ID, verificamos:

#### Chequeamos que la cantidad de valores es igual a la cantidad de elementos index del data frame

```python
len(ds.iloc[:,0].unique()) == len(ds.index)
```

#### Chequeamos que realmente cada elemento de la columna 0 tenga un elemento igual en el index, preguntando por cada valor en la columna 0 si NO está en el index. Si es que hay una relación 1 a 1 entre el index y los valores de la columna 0, entonces esta consulta tiene que retornar 0 rows diferentes

```python
rowsDiferentes = ds.loc[~ds.iloc[:,0].isin(ds.index)]
len(rowsDiferentes) == 0
```

### Definimos una funcion para obtner información básica acerca de una columna

```python
def getColumnBasicInfo(dataframe, columnName):
    info = '';
    uniqueValuesCount = len(dataframe[columnName].unique())
    
    info = info + 'Data type: ' + str(type(dataframe[columnName].unique()[0])) + '\n'
    
    info = info + 'Count TOTAL values: ' + str(len(dataframe[columnName])) + '\n'
    
    info = info + 'Count UNIQUE values: ' + str(uniqueValuesCount) + '\n'
    
    if(uniqueValuesCount < 30):
        info = info + 'Unique values: ' + str(dataframe[columnName].unique()).strip('[]') + '\n'
    else:
        info = info + 'Unique values: ' + str(dataframe[columnName].unique()[0:30]).strip('[]') + '....ETC....' + '\n'
    
    
    hasNullValues = dataframe[columnName].isna().any()
    
    info = info + 'Has NULL values: ' + str(hasNullValues) + '\n'
    
    if(hasNullValues):
        info = info + 'NULL values count: ' + str(dataframe[columnName].isna().sum())
    
    print(info)
       
```
